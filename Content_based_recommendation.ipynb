```python
indices = pd.Series(index=movies_data['original_title'], data=movies_data.index)
indices.head()
```

```python
print(movies_data['overview'].isnull().sum())
movies_data['overview'].fillna(' ', inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer(ngram_range=(1,3),stop_words='english', lowercase=True, min_df=3)
tf_idf_matrix = tf_idf.fit_transform(movies_data['overview'])
print(tf_idf_matrix.shape)
```

```python
tf_idf_matrix_1 = pd.DataFrame(data=tf_idf_matrix.toarray(), columns=tf_idf.get_feature_names())
tf_idf_matrix_1
```

```python
# Lets See tfidf values of movie at index 99
print('Movie Name :',movies_data.iloc[99]['original_title'])
print('OverView :', movies_data.iloc[99]['overview'])
print("--------------------------------------------")
for index , value in tf_idf_matrix_1.iloc[99].items():
    if value>0:
        print(index , ":", value)
```

```python
# Compute Sigmoid Kernal to get the relation value of one movie with every other movie
from sklearn.metrics.pairwise import sigmoid_kernel
sigmoid = sigmoid_kernel(tf_idf_matrix, tf_idf_matrix)
sigmoid.shape
```

```python
print("Index of Movie locker 13 in original  dataset is :",indices['Locker 13'])
print('Similarity rating of movies with Locker 13 based on overview ',(sigmoid[indices['Locker 13']]))
```

```python
# Content_based_recommendations
def get_content_based_recommendation(movie_name):
    movie_index = indices[movie_name]
    sigmoid_value = sorted(list(enumerate(sigmoid[indices[movie_name]])), key=lambda x : x[1], reverse=True)
    sigmoid_value = sigmoid_value[1:11]
    original_index = [index[0] for index in sigmoid_value]
    df = movies_data.iloc[original_index][['original_title','vote_count','vote_average']]
    return df
df = get_content_based_recommendation('The Dark Knight')
df
```
